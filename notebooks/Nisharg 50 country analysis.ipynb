```python
# Group by the 'Country' column
grouped = countries_of_interest_df.groupby('Country')

# Find countries with zero missing values throughout all the years
countries_with_no_missing_values = []

for country, group in grouped:
    if group.isna().sum().sum() == 0:
        countries_with_no_missing_values.append(country)

print("Countries with zero missing values throughout all the years:", countries_with_no_missing_values)
```

```python
# Calculate the percentage of null values for each column
null_percentage = countries_of_interest_df.isna().mean() * 100

# Plot the percentage of null values for each column
plt.figure(figsize=(12, 8))
null_percentage.plot(kind='bar', color='skyblue')
plt.title('Percentage of Null Values by Column')
plt.xlabel('Columns')
plt.ylabel('Percentage of Null Values (%)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Calculate the percentage of null values for each column
null_percentage = countries_of_interest_df.isna().mean() * 100

# Identify columns with more than 60% missing data
columns_with_high_nulls = null_percentage[null_percentage > 60].index

# Drop these columns from the DataFrame
cleaned_df = countries_of_interest_df.drop(columns=columns_with_high_nulls)

# Output the names of the columns that were dropped
print("Columns with more than 60% missing data:")
print(columns_with_high_nulls)

# Calculate the percentage of null values for each column in the cleaned DataFrame
cleaned_null_percentage = cleaned_df.isna().mean() * 100

# # Print the cleaned DataFrame
# print("Cleaned DataFrame:")
# print(cleaned_df)

# Plot the null value percentage of the cleaned DataFrame
plt.figure(figsize=(12, 8))
cleaned_null_percentage.plot(kind='bar', color='skyblue')
plt.title('Percentage of Null Values by Column (Cleaned DataFrame)')
plt.xlabel('Columns')
plt.ylabel('Percentage of Null Values (%)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Group by 'Country' and find the first non-null 'Year' and the last non-null 'Year' for each country
first_years = cleaned_df.groupby('Country')['Year'].min()
last_years = cleaned_df.groupby('Country')['Year'].max()

# Create a sorted list of countries for consistent plotting
sorted_countries = first_years.index

# Plot the starting and ending year of data for each country
plt.figure(figsize=(12, 8))
plt.scatter(sorted_countries, first_years[sorted_countries], color='skyblue', label='Starting Year')
plt.scatter(sorted_countries, last_years[sorted_countries], color='orange', label='Ending Year')

# Connect starting and ending years with a line
for country in sorted_countries:
    plt.plot([country, country], [first_years[country], last_years[country]], color='gray', linestyle='--')

plt.title('Starting and Ending Years of Data for Each Country')
plt.xlabel('Country')
plt.ylabel('Year')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.legend()

# Adding data labels to each point
for i, country in enumerate(sorted_countries):
    plt.text(i, first_years[country], str(first_years[country]), ha='center', va='bottom', fontsize=8)
    plt.text(i, last_years[country], str(last_years[country]), ha='center', va='bottom', fontsize=8)

plt.show()
```

```python
grouped = cleaned_df.groupby('Country')

# Initialize a list to store countries and the years after which there are no missing values
no_missing_values_years = []

# Iterate over each country
for country, data in grouped:
    # Drop rows with any missing values
    no_missing_values_data = data.dropna(axis=0, how='any')
    
    # Check if there are any rows left after dropping missing values
    if not no_missing_values_data.empty:
        # Find the earliest year after which there are no missing values
        earliest_year = no_missing_values_data['Year'].min()
        
        # Append the country and the earliest year to the list
        no_missing_values_years.append((country, earliest_year))

# Print the list of countries and the years after which there are no missing values
print("List of countries and the years after which there are no missing values in all columns:")
for country, year in no_missing_values_years:
    print(f"{country}: {year}")
```

```python
import pandas as pd
from tabulate import tabulate

# Assuming your DataFrame is named cleaned_df

# Group by country
grouped = cleaned_df.groupby('Country')

# Initialize dictionaries to store starting and ending years and the year after which there are no missing values
starting_ending_years = {}
no_missing_values_years = {
    "Armenia": 2005,
    "Azerbaijan": 2000,
    "Belarus": 2004,
    "Bosnia and Herzegovina": 1999,
    "Croatia": 1998,
    "Czech Republic": 1996,
    "Estonia": 1995,
    "Georgia": 2000,
    "Kazakhstan": 2002,
    "Kyrgyz Republic": 2000,
    "Latvia": 1998,
    "Moldova": 1995,
    "Russia": 1998,
    "Serbia": 2001,
    "Slovak Republic": 1995,
    "Slovenia": 1995,
    "Sri Lanka": 1990,
    "Ukraine": 1997,
    "Uzbekistan": 2002
}

# Iterate over each country
for country, data in grouped:
    # Find the starting and ending years
    starting_year = data['Year'].min()
    ending_year = data['Year'].max()
    starting_ending_years[country] = (starting_year, ending_year)

# Initialize a list to store combined data
combined_data = []

# Merge the data
for country, (starting_year, ending_year) in starting_ending_years.items():
    if country in no_missing_values_years:
        no_missing_year = no_missing_values_years[country]
    else:
        no_missing_year = None
    combined_data.append([country, starting_year, ending_year, no_missing_year])

# Format the list into a table
table = tabulate(combined_data, headers=['Country', 'Starting Year', 'Ending Year', 'Year after which there are no missing values'], tablefmt='grid')

# Output the formatted table
print("List of countries and their starting and ending years of data, and the year after which there are no missing values (sorted by starting year):")
print(table)
```
